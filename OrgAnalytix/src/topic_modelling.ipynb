```python
import pandas as pd
import numpy as np
from nltk import word_tokenize
from nltk.stem.snowball import SnowballStemmer
import gensim
from gensim.models import Doc2Vec
from gensim.models.doc2vec import TaggedDocument
from collections import namedtuple
import warnings
warnings.filterwarnings("ignore",category=DeprecationWarning)
import multiprocessing
cores = multiprocessing.cpu_count()
```

```python
filepath = '../data/quotes_draft.csv'
```

```python
quotes_df = pd.read_csv(filepath, index_col = 0)
```

```python
quotes_df.head()
```

```python
def tagged_doc_from_df(df):
    '''turns rows in a dataframe into gensim-compatible taggeddocs'''
    docs = []
    for i in df.iterrows():
        if type(i[1]['quote']) is not str:
            pass
        elif len(i[1]['quote'].split(' ')) <= 3:
            pass
        else:
            d = TaggedDocument(word_tokenize(i[1]['quote']), [(i[1]['char1'],i[1]['char2'])])
            docs.append(d)
    return docs
```

```python
taggeddocuments = tagged_doc_from_df(quotes_df)
```

```python
model1 = Doc2Vec(taggeddocuments, dm=1, dm_concat=1, vector_size=100, window=3, negative=3, hs=0, min_count=2, workers=cores, epochs = 100)
```

```python
model2 = Doc2Vec(taggeddocuments,dm=0, vector_size=100, negative=3, hs=0, min_count=2, workers=cores,epochs = 100)
```
